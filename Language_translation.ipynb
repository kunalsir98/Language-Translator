```python
import os 
from dotenv import load_dotenv
load_dotenv()

groq_api_key=os.getenv("GROQ_API_KEY")
#groq_api_key 
```

```python
from langchain_groq import ChatGroq
model=ChatGroq(model="llama-3.3-70b-versatile",groq_api_key=groq_api_key)
model
```

```python
from langchain_core.messages import SystemMessage,HumanMessage
messages=[
    SystemMessage(content="Translate this from English to Hindi"),
    HumanMessage(content="Hello my name is kunal")
]

result=model.invoke(messages)
result
```

```python
from langchain_core.output_parsers import StrOutputParser
str_parser=StrOutputParser()
str_parser.invoke(result)
```

```python
chain=model|str_parser
chain.invoke(messages)
```

## Prompt Templating 

```python
from langchain_core.prompts import ChatPromptTemplate

generic_template="Translate the following message{language}:"

prompt=ChatPromptTemplate.from_messages([
    ("system", generic_template),
    ("user","{text}")
])
```

```python
result=prompt.invoke({"language":"Hindi","text":"Hello my name is kunal and I am a Gen AI Developer"})
```

```python
result.to_messages()
```

```python

chain=prompt|model|str_parser
chain.invoke({"language":"Hindi","text":"Hello my name is kunal and I am a Gen AI Developer"})
```
